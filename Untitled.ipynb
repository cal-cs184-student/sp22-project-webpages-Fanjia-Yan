```python
import numpy as np
from matplotlib import pyplot as plt
import random
import math
```

```python
def hop_flea(x):
    N = len(x)
    x_0 = random.randint(0,N-1);
    if x[x_0] == 1:
        x[x_0] = 2
    else:
        x[x_0] = 1
    return x
```

```python
N = 200;
T = 20*N;

flea = np.ones(N)
log_val_list = []
time = np.arange(0,T)

for i in range(T):
    flea = hop_flea(flea)
    dog_1_count = np.count_nonzero(flea == 1)
    log_count = math.log(dog_1_count)
    log_val_list.append(log_count)   
plt.plot(time,log_val_list)
```
